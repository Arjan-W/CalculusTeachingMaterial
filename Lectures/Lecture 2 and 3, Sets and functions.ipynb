# Sets
A set is a collection of elements where each elements without duplicates. This means that a set cannot contain the same element twice. For example, $\{1,2,3,a,b\}$ denotes a set but $\{1,1,2\}$ does not as the element $1$ is duplicated.

Sets can be either finite or infinite. Infinite sets contain a infinite numbers of elements.

In this course we are mainly interested in sets of numbers. The most useful examples are:

### The set of natural numbers
This is the set of numbers you learned to use in primary school. In this course we will consider $0$ to be a natural number.

$$
N = \{0, 1, 2, 3, 4, ...\}
$$

### The set of integers
Integers are natural numbers with a sign. Each natural number except for $0$ has its negative version:

$$
Z = \{..., -2, -1, 0, 1, 2, ...\}
$$

We will denote the positive integers as:

$$
Z^{+} = \{1, 2, 3, 4, ...\}
$$

and the negative integers as:

$$
Z^{-} = \{-1, -2, -3, ...\}
$$

### The set of rational numbers
Rational numbers are defined as ratios of integers:
$$
q = \frac{a}{b} 
$$
where $a$ is a integer and $b$ is a positive integers. We denote the set of rational numbers as $Q$.

Note that two fractions can represent the same rational number. For example:

$$
\frac{1}{2} = \frac{2}{4}
$$

### The set of real numbers
Not all numbers can be expressed as a ratio of integers. The simplest example is the square root of two: $\sqrt{2}$.

Numbers that are not rational are said to be irrational. The set of eal numbers contains both rational and irrational numbers. A real number is defined by a (usually infinite) sequence of digits. For example, the square root of two in base $10$ is:

$$
\sqrt{2} = 1.41421356237309504880168...
$$

Note that two sequences of digits can represent the same number, for example:

$$
1.00000000000... = 0.999999999... 
$$

real numbers can be decomposed into a fractional part and a integral part, for example:

$$
\sqrt{2} = 1 + 0.41421356237309504880168...
$$

#### Formal definition 
More formally, the fractional part of a real number can be defined as the limit of a infinite sum. Consider a infinite sequence of digits:
$$
A = A_1, A_2, A_3, ...
$$
from these digits, we can construct a real number $a$ as the following infinite sum of rational numbers
$$
x = \sum_{n=1}^\infty \frac{A_n}{10^n}~.
$$
Note that the division by a power of ten places the digits in the right position in the decimal expansion:
$$
\begin{aligned}
&0.\color{blue}{4}\color{red}{1}\color{green}{4}\color{purple}{2}... = \\
&0.\color{blue}{4}000... + \\
&0.0\color{red}{1}00... + \\
&0.00\color{green}{4}0... + \\
&0.000\color{purple}{2}... + \\
\end{aligned}
$$

using this definition, it is easy to see that $1.0000000000...$ = $0.9999999999$:

$$
1.00000... = 1 + \sum_{n=1}^\infty \frac{0}{10^n} = 1
$$

$$
0.99999... = 0 + \sum_{n=1}^\infty \frac{9}{10^n} = 9 \sum_{n=1}^\infty (0.1)^n = 9 \left(\sum_{n=0}^\infty (0.1)^n - 1 \right) = 9 \left(\frac{1}{1 - 0.1} - 1 \right) = 1
$$
where I used the formula for summing the infinite geometric series (see lecture 1).

# Functions

In this lecture we are going to discuss functions. A function between the set $A$ and a set $B$ is defined as a "rule" that assigns one and only one element of $B$ for each element of $A$.

We denote a function from $A$ to $B$ as follows:

$$
f:A \rightarrow B
$$

For each element $a$ of $A$, we denote the corresponding element of $B$ as $f(a)$. The (input) set $A$ is called domain while the (output) set $B$ is called codomain.

The following figure visualize a function from the set (of characters) $A = \{a, b, c\}$ to the set (of numbers) $B = \{1, 2\}$.

A function can be seen as a (potentially infinite) table of pairs. For example, the function in the image is defined by the following table:
$$
\begin{aligned}
&(a, f(a)) = (a, 1) \\
&(b, f(b)) = (b, 2) \\
&(c, f(c)) = (c, 2)
\end{aligned}
$$

## Injective functions

A function $f:A \rightarrow B$ is said to be injective if, for all different $a_1$ and $a_2$ in $A$, $f(a_1)$ is always different from $f(a_2)$. In other words, a injective function never maps to elements of $A$ on the same element of $B$.

The function in the previous figure is not injective as $f(b) = f(c) = 2$.

## Surjective functions

The range (or image) of a function is the set of its values. We will denote the range of $f$ as $f(A)$:

$$
f(A) = \{f(a) ~~|~~ \text{for all a in A}\}
$$

Clearly, the range of a function $f: A \rightarrow B$ is a subset of $B$. The function $f$ is said to be surjective when its range is identical to B. This means that all elements of $B$ can be written as $f(a)$ for some $a$ in $A$.

The function in the figure is surjective as $f(A) = \{f(a), f(b)\} = \{1, 2\} = B$.

## One-to-one functions and their inverse function

A function is said to be one-to-one when it is both injective and surjective. This means that each element $b$ of $B$ can be written as $f(a)$ in a unique way. 
When a function $f:A \rightarrow B$ is one-to-one, it is possible to obtain a inverse function $f^{-1}: B \rightarrow A$ such that:

$$
f^{-1}(f(a)) = a
$$

and

$$
f(f^{-1}(b)) = b ~.
$$

The inverse function is obtained simply by reversing the order of the pairs. For example, the inverse of:

$$
\begin{aligned}
&(1, f(1)) = (1, 1) \\
&(2, f(2)) = (2, 3) \\
&(3, f(3)) = (3, 2)
\end{aligned}
$$

is 

$$
\begin{aligned}
&(1, f^{-1}(1)) = (1, 1) \\
&(3, f^{-1}(3)) = (3, 2) \\
&(2, f^{-1}(2)) = (2, 3)
\end{aligned}
$$

Note that this is possible only if the function is one-to-one since: 1) $f^{-1}(b)$ needs to be defined for all $b$ in $B$, this is possible only if $f$ is surjective. 2) $f^{-1}(b)$ has to be uniquely defined. This is only true when $f$ is injective.

### Examples

- The function $f_1(x) = x$ is equal to its own inverse.
- The inverse of $f_2(x) = x + 1$ is $f_2^{-1}(x) = x - 1$.
- The inverse of $f_3(x) = 3 x$ is $f_3^{-1}(x) = x / 3$

## Linear functions

linear functions are obtained by multiplying the input by a factor:

$$
f(x) = a x
$$

If the factor $a$ is equal to zero, the function is the $f(x) = 0$ constant function. Otherwise, the linear function is one-to-one and its inverse is given by:

$$
f^{-1}(x) = x/a
$$

note that the inverse of a linear function is itself a linear function. Here is the plot of the function $f(x) = 2 x$ and its inverse:

```python
f = lambda x: 2*x
f_inv =  lambda x: x/2

x_range = np.linspace(-1,1,100)
p = 3

plt.plot(x_range, [f(x) for x in x_range], label="f(x) = 2x")
plt.plot(x_range, [f_inv(x) for x in x_range], label="f^-1(x) = x/2")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

## Powers and roots

Powers functions are obtained by taking a integer power of the input:

$$
f_p(x) = x^p = \underbrace{x \cdot x \cdot x ... \cdot x}_\text{$p$ times}
$$

When the power $p$ is odd, the function is one-to-one and it can be inverted. Conversely, when $p$ is even, the function is one-to-one only if we restrict the domain to non-negative real numbers. The inverse is given by the $p$-th root function:

$$
f_p^{-1}(x) = \sqrt[p]{x}
$$

Note that, when $p$ is even, $\sqrt[p]{x}$ is defined only when $p$ is non-negative.

We can now plot the graph of some power functions:

```python
f = lambda x,p: x**p

for p in range(6):
    plt.plot(x_range, f(x_range, p), label="f(x) = x^{}".format(p))
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

and of the corresponding root functions:

```python
def f_inv(x, p):
    if x>= 0:
        return x**(1/p)
    else:
        if p%2 == 0:
            return np.nan
        else:
            return -(abs(x))**(1/p)
        
for p in range(1,5):
    plt.plot(x_range, [f_inv(x, p) for x in x_range], label="f(x) = x^{}".format(p))
plt.xlim(min(x_range), max(x_range))
plt.ylim(-1.5, 1.5)
plt.legend(loc="best")
```

Root functions can also be written as fractional roots:

$$
\sqrt[p]{x} = x^{1/p}
$$

In fact, since roots are the inverse functions of powers, we have that:

$$
(x^p)^{1/p} = x^{p/p} = x^1 = x

## Power series

Power series are defined as a infinite weighted sum of powers:

$$
f(x) = \sum_{n=0}^\infty \beta_n x^n
$$

where the $\beta_p$ form a infinite sequence of coefficients. loosely speaking, a power series is a kind of polynomial with infinite degree.

Given a real number $x$, the value $f(x)$ is defined as the usual limit of partial sums:

$$
f(x) = \lim_{N \rightarrow \infty} \sum_{n=0}^N \beta_n x^n
$$

Note that this limit does not always exist. The domain of the power series is then defined as the set containing all the values of $x$ for which the limit exists.

This will become clearer with a example. Consider the series defined by the infinite geometric summed we discussed in the previous lecture:

$$
f(x) = \lim_{N \rightarrow \infty} \sum_{n=0}^N x^n
$$

From the previous lecture, we know that if the absolute value of $x$ is smaller than $1$:

$$
f(x) = \lim_{N \rightarrow \infty} \sum_{n=0}^N x^n = \lim_{N \rightarrow \infty} \frac{1 - x^{N+1}}{1 - x} = \frac{1}{1 - x}
$$

Therefore, the domain of this power series is the inteval $(-1, 1)$. The following graph shows this function together with a series of approximations obtained by summing only finitly many powers:

## The exponential function

The exponential function is one of the most important funnction that we will discuss in this course. The exponential function is defined by the following infinite series:

$$
\exp{(x)} = \sum_{n=0}^\infty \frac{x^n}{n!}
$$

where $n!$ is the factorial of $n$:

$$
n! = n \cdot (n-1) \cdot (n-2) ... \cdot 2 \cdot 1~.
$$

It can be shown that this infinite series converges for all values of $x$ (the reason is that $n!$ goes to zero very quickly and makes the high order terms smaller and smaller.). Therefore, the domain of the exponential is the whole set of real numbers.

### Properties of the exponential function
The esponential function is positive-valued and its domain is the whole set of real numbers. The most important property of the exponential is that it converts sums into products:

$$
\exp(a+b) = \exp(a)\exp(b)
$$

From this property, we can see that:

$$
\exp(x) = \exp(0 + x) = \exp(0)\exp(x)
$$

Hence:

$$
\exp(0) = 1
$$

Furthermore:

$$
1 = \exp(0) = \exp(x + (- x)) = \exp(x)\exp(-x)
$$

Hence:

$$
\exp(-x) = 1/\exp(x)~.
$$

#### Bonus section for hard core students
We can prove $\exp(a+b) = \exp(a)\exp(b)$ using the definition (and the binomial theorem). let's start by writing the exponentials as infinite sums:

$$
\exp(a)\exp(b) = \left(\sum_{k=0}^\infty \frac{a^k}{k!}\right) \left(\sum_{m=0}^\infty \frac{b^m}{m!}\right)
$$

We can write the product of these two sums as a double sum over the pairwise multiplied terms:

$$
\left(\sum_{k=0}^\infty \frac{a^k}{k!}\right) \left(\sum_{m=0}^\infty \frac{b^m}{m!}\right) = \sum_{k=0}^\infty \sum_{m=0}^\infty \frac{a^k b^m}{k! m!}
$$

Let's now consider the sum of all the terms such that $k + m$ is equal to some fixed integer $n$:

$$
\sum_{k=0}^n \frac{a^k b^{n - k}}{k! (n-k)!}
$$
here we wrote $m$ as $n - k$ since $m + k = 1$.

We can write our original double sum by summing all these finite sums for all possible values of $n$:

$$
\sum_{k=0}^\infty \sum_{m=0}^\infty \frac{a^k b^m}{k! m!} = \sum_{n=0}^\infty \left( \sum_{k=0}^n \frac{a^k b^{n - k}}{k! (n-k)!} \right) 
$$

from the binomial theorem, we know that:

$$
(a+b)^n = \sum_{k=0}^n \frac{n!}{(n-k)! k!} a^n b^{n - k}
$$

Therefore, the inner term in our sum is

$$
\sum_{k=0}^n \frac{a^n b^{n - k}}{k! (n-k)!} = \frac{(a+b)^n}{n!}~.
$$

by plugging this into our double sum we get the final result:

$$
\exp(a)\exp(b) = \sum_{n=0}^\infty \left( \sum_{k=0}^n \frac{a^k b^{n - k}}{k! (n-k)!} \right) = \sum_{n=0}^\infty \frac{(a+b)^n}{n!} = \exp(a + b)
$$

Q.E.D.

### Exponential function with integer inputs

Let's consider the case where the input of the exponential function is a natural number:

$$
\exp(n)
$$

Using the property of the exponential, we can derive a simple recursive relation:

$$
\begin{aligned}
\exp(n) &= \exp(n - 1) \exp(1) \\
&= \exp(n - 2) \exp(1) \exp(1) \\
&= \underbrace{\exp(1) \cdot \exp(1) \cdot ... \cdot \exp(1)}_\text{$n$ times} \\
&= \exp(1)^n
\end{aligned}
$$

Thereforem when $n$ is a integer the exponential can be written as the number $\exp(1)$ to the $n$-th power. This number is called the Euler constant (denoted as $e$) and it is defined by the following infinite sum:

$$
e = \exp(1) = \sum_{n=1}^\infty \frac{1}{n!}
$$

Therefore, we can write $\exp(n)$ as

$$
e^n
$$

Similarly, we will sometime write

$$
\exp(x) = e^x~.
$$

note that this is just a convenient notation as it does not make sense to say that the number $e$ should be multiplied by itself $x$ times when $x$ is not a integer.

### The natural logarithm

The natural logarithm is the inverse function of the exponential:

$$
\log(\exp(x)) = \exp(\log(x)) = x
$$
Since the exponential is positive-valued, the domain of the logarithm is the set of positive real numbers. In other words, we cannot define the logarithm of a negative number (or of zero) as the value of the exponential is always positive.

Being the inverse of the exponential, the logarithm converts products into sums:

$$
\log(a \cdot b) = \log(a) + \log(b)
$$

The logarithm also converts ratios into differences:

$$
\log(a/b) = \log(a) - \log(b)
$$


### Exponential functions with different bases

There are infinitely many other version of the exponential function all respecting the property:

$$
y^a y^b = y^{a+b}
$$

when $a$ and $b$ are integers these functions are easy to define:

$$
y^a = \underbrace{y \cdot y \cdot ... \cdot y}_\text{$a$ times}
$$

we now need to define $y^x$ when $x$ is a arbitrary real number. We can define this using the exponential function and the logarithm. In fact, the number $y$ can be written as follows:

$$
y = \exp(\log(y))~.
$$

Hence:

$$
y^x = \exp(\log(y))^x = (e^{\log(y)})^x = e^{x \log(y)}
$$

```python
from scipy.special import factorial 
x_range = np.linspace(-4,4,100)

f = lambda x,N: sum([x**n/factorial(n) for n in range(N)])

for N in range(1,10, 2):
    plt.plot(x_range, [f(x, N) for x in x_range], label="N = {}".format(N-1))
plt.plot(x_range, np.exp(x_range), c="k", lw=2, label="exp(x)")
plt.xlim(-4,4)
plt.ylim(-1,4)
plt.legend(loc="best")
plt.title("Exponential function and truncation of its power series")
```

## Trigonometric functions

The trigonometric functions are strictly connected with circles. The equation for a circle centered at the origin with radius one is:

$$
\color{blue}{x^2} + \color{green}{y^2} = 1~.
$$

Each point in the circle is uniquely defined by a angle $\color{purple}{\theta}$ ranging from $0$ to $2 \pi$ radians. Points in the circle can then be expressed as function of the angle using the cosine and the sine function:

$$
\color{blue}{\cos(\color{purple}{\theta})^2} + \color{green}{\sin(\color{purple}{\theta})^2} = 1~.
$$


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/Sinus_und_Kosinus_am_Einheitskreis_1.svg/500px-Sinus_und_Kosinus_am_Einheitskreis_1.svg.png" alt="drawing" width="300"/>

Therefore, the cosine gives the $x$ coordinate of a point in a circle of radius one while the sine gives the $y$ coordinate.

From the diagram is clear that, when $\color{purple}{\theta}$ is equal to zero, the sine is equal to zero while the cosine is equal to one. Conversely, the cosine is zero and the sine is one when $\color{purple}{\theta}$ is equal to $\pi/2$.

The result are functions that oscillate up and down:

```python
x_range = np.linspace(0,2*np.pi,300)

plt.plot(x_range, np.sin(x_range), lw=2, label="sin(x)", c="g")
plt.plot(x_range, np.cos(x_range), lw=2, label="cos(x)", c="b")
plt.xlim(0,2*np.pi)
plt.legend(loc="best")
plt.title("Trigonometric functions")
```

### Important trigonometric identities

There are many important identities relating the values of the trigonometric functions. You can find a comprehensive list [here](https://en.wikipedia.org/wiki/List_of_trigonometric_identities).

We already introduced the most important identity:

$$
\cos(x)^2 + \sin(x)^2 = 1 ~.
$$

As you can see from their graphs, sine and cosine can be obtained from each other by shifting the input:

$$
\sin(x) = \cos(x + \pi/2)
$$

$$
\cos(x) = \sin(x - \pi/2)
$$

Other two very important identities can be used for doubling the angle:

$$
\sin(2x) = 2\cos(x)\sin(x)
$$

$$
\cos(2x) = \cos(x)^2 - \sin(x)^2
$$

These are special cases of formulas for summing angles:

$$
\sin(x + y) = \cos(x)\sin(y) + \cos(y)\sin(x)
$$

$$
\cos(x + y) = \cos(x)\cos(y) - \sin(y)\sin(x)
$$

We can also invert the angle doubling formula to get angle halfing formulas:

$$
\sin(x/2)^2 = \frac{1 - \cos(x)}{2}
$$

$$
\cos(x/2)^2 = \frac{1 + \cos(x)}{2}
$$

The power series expressions that we just gave look suspitiosly similar to the power series of the exponentional function. This is not a conincidence! In this section we will see that sine and cosine functions are just different manifestations of the exponential function!

The section requires the use umaginary numbers. Don't worry though, it is optional and you need to know a single new formula!
Here is the formula, we introduce a new (imaginary) number $i$ whose square is equal to $-1$:

$$ 
i^2 = -1
$$

Imaginary numbers have the following form:

$$
i \color{green}{b}
$$

where $\color{green}{b}$ is an arbitrary real number. A genal complex number is obtained by summing a real and an imaginary number:

$$
\color{blue}{a} + i \color{green}{b}
$$

Let's now try to input a immaginary number to the exponential function:

$$
\exp{(i x)} = \sum_{n=0}^\infty \frac{(i x)^n}{n!}
$$

We can split this sums into two terms, one involving all the even numbers ($2 n$) and the other involving all the odd numbers ($2n + 1$):

$$
\sum_{n=0}^\infty \frac{(i x)^n}{n!} = \sum_{n=0}^\infty \frac{(i)^n x^n}{n!} = \underbrace{\sum_{n=0}^\infty \frac{(i)^{2n} x^{2n}}{(2n)!}}_\text{Even numbers} + \underbrace{\sum_{n=0}^\infty \frac{(i)^{2n+1} x^{2n+1}}{(2n+1)!}}_\text{Odd numbers}
$$

let's now simplify the even part. It jeasy to give a simple expression for the imaginary number $i$ raised to a even power $2n$:

$$
i^{2n} = (i^2)^n = (-1)^n
$$

Therefore, $i^(2n)$ oscillates between plus and minus one. Using this formula, we get:

$$
\sum_{n=0}^\infty \frac{(i)^{2n} x^{2n}}{(2n)!} = \color{blue}{\sum_{n=0}^\infty \frac{(-1)^{n} x^{2n}}{(2n)!}} = \cos{(x)}
$$

Where the series in blue is our previous definition of the series of the cosine! 

Similarly:

$$
i^{2n+1} = i (i^2)^n = i (-1)^n
$$

and

$$
\sum_{n=0}^\infty \frac{(i)^{2n+1} x^{2n+1}}{(2n+1)!} = \sum_{n=0}^\infty \frac{i(-1)^{n} x^{2n}}{(2n)!} = i \color{green}{\sum_{n=0}^\infty \frac{(-1)^{n} x^{2n}}{(2n)!}} = i \sin{(x)}~.
$$

Putting these two expression together, we obtain a very important identity:

$$
\exp(i x) = \color{blue}{\cos(x)} + i \color{green}{\sin(x)}~.
$$

In words, the cosine is the real part of the imaginary exponential while the sine is its imaginary part. 

We can now derive one the most beautiful formula of mathematics as special case! Let's input $pi$ to this expression:

$$
\exp(i \pi) = \cos(\pi) + \sin(\pi) = -1 + 0
$$

The result is the very famous Euler identity:

$$
e^{i \pi} + 1 = 0
$$

## The tangent and the inverse trigonometric functions

The tangent is defined as the ration between sine and cosine:

$$
\tan{(x)} = \frac{\sin(x)}{\cos(x)}
$$

Clearly, this expression makes sense only when the denominator is different from zero (you can't divide by zero!). Therefore, the domain of the tangent is the set of real numbers $z$ such that $\cos(z) \neq 0$.

```python
x_range = np.linspace(-2*np.pi,2*np.pi,1000)

f = lambda x: np.tan(x) if np.abs(np.cos(x))>0.01 else np.nan

plt.plot(x_range, [f(x) for x in x_range], lw=2, label="tan(x)", c="g")
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-5,5)
plt.legend(loc="best")
plt.title("Tangent")
```

As you can see from the graph, the tangent is periodic with period $\pi$.

### Inverse functions

Sine, cosinge and tangents are periodic and therefore they are clearly not one-to-one. This means that we cannot define their inverses. 

For example, there are infinitly many values of $x such that:

$$
\cos(x) = 1
$$

In particular, we have that:

$$
\cos(n 2 \pi) = 1
$$

for all possible integers $n$.

#### The arcosine function
However, can can define the inverse functions of the restriction of the trigonometric functions to a smaller domain. For example, consider the sine function defined between $-\pi/2$ and $\pi/2$:

```python
x_range = np.linspace(-np.pi/2,np.pi/2,300)

plt.plot(x_range, np.sin(x_range), lw=2, label="sin(x)", c="g")
plt.xlim(-np.pi/2,np.pi/2)
plt.legend(loc="best")
plt.title("restriction of the sine function")
```

As you can see from the graph, the sine is one-to-one in this interval. This imply that we can define the inverse function of this restriction.

We denote the inverse of the sine between $-\pi/2$ and $\pi/2$ as:

$$
\arcsin(x)
$$

Sine the values of the sine are between $-1$ and $1$, the dimain of the arcosine is the interval between $-1$ and $1$. Here is its graph

```python
x_range = np.linspace(-1,1,300)

plt.plot(x_range, np.arcsin(x_range), lw=2, label="arcsin(x)", c="g")
plt.xlim(-1,1)
plt.legend(loc="best")
plt.title("Arcosine function")
```

#### The arcocosine function

Similarly, we can define the arcocosine function by restrincting the cosine in the interval between $0$ and $\pi$:

```python
x_range = np.linspace(0,np.pi,300)

plt.plot(x_range, np.cos(x_range), lw=2, label="cos(x)", c="g")
plt.xlim(0,np.pi)
plt.legend(loc="best")
plt.title("Restriction of the cosine function")
```

```python
x_range = np.linspace(-1,1,300)

plt.plot(x_range, np.arccos(x_range), lw=2, label="arccos(x)", c="g")
plt.xlim(-1,1)
plt.legend(loc="best")
plt.title("Arcocosine function")
```

#### The arcotangent function

Finally, we can define the arcotangent function by inverting the restriction of the tangent in the interval between $-\pi/2$ and $\pi/2$. Note that $\tan(x)$ tends to plus infinity when $x$ tends to $/pi/2$ and to minus infinity when $x$ tends to $-\pi/2$.

```python
x_range = np.linspace(-np.pi/2,np.pi/2,300)

plt.plot(x_range, np.tan(x_range), lw=2, label="tan(x)", c="g")
plt.xlim(-np.pi/2,np.pi/2)
plt.ylim(-10,10)
plt.legend(loc="best")
plt.title("restriction of the tangent function")
```

The arcotangent function is the inverse of the tangent in this interval. Its domain is the whole set of real numbers since tangent assumes all possible values.

```python
x_range = np.linspace(-10,10,300)

plt.plot(x_range, np.arctan(x_range), lw=2, label="arctan(x)", c="g")
plt.xlim(-10,10)
plt.legend(loc="best")
plt.title("Arcotangent function")
```

# Limits of a function



```python
f = lambda x: x**2

x_range = np.linspace(-3,3,100)
plt.plot(x_range, f(x_range), c="k")
a = 1
plt.scatter(a, f(a), c="r")
epsilon = 0.5
plt.fill_between(x_range, f(a)-epsilon, f(a)+epsilon, alpha = 0.2, color = "b", label = "|f(x) - l| < epsilon")
delta = 0.2
plt.axvspan(a - delta, a + delta, alpha=0.2, color='red', label = "|x - a| < delta")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

## Limits at infinity

```python
f = lambda x: x**2/(1 + x**2)

x_range = np.linspace(0,10,100)
plt.plot(x_range, f(x_range), c="k")
l = 1
plt.axhline(l, c = "k", lw = 2, ls = "--", label="Limit")
epsilon = 0.05
plt.fill_between(x_range, l-epsilon, l+epsilon, alpha = 0.2, color = "b", label = "|f(x) - l| < epsilon")
X = 5
plt.axvspan(X, max(x_range), alpha=0.2, color='red', label = "x > X")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```
